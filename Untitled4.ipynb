```python
# Solve the data that has "?"
sol = df[df['F'] != '?']['F'].astype(int).mean()
df['F'] = df['F'].replace('?', sol).astype('int64')
```

```python
# Correction Matrix
sns.set(style = 'dark', color_codes = True)
sns.set(font_scale = 1.5)
plt.figure(figsize = (16,10))
sns.heatmap(df.corr(), cmap='viridis', annot = True)
```

4. Choose features to be included in your model(explain your reason)

After reviewing the data and creating the correlation matrix, the characteristics "A," "B," "C," "D," "E," "F," "G," and "H" are used. With the exception of "I," which has a correlation of only 0.42, all of these characteristics have a strong correlate with cancer. When added to the model, the "I" feature may cause overfitting and no accuracy increase. As a result, I decided to remove the "I" component from my model.

```python
used_feature = ["A","B","C","D","E","F","G","H"]
x = df[used_feature]
y = df["Cancer"]
```

```python
# Splitting Data set
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
# Normalize input data
means = np.mean(x_train,axis = 0)
stds = np.std(x_train, axis = 0)

x_train = (x_train-means)/stds
x_test = (x_test-means)/stds
```

```python
# Logistic Regression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_LogisticPred = model.predict(x_test)
```

```python
# Model Performance
ALP = accuracy_score(y_test,y_LogisticPred)
print("Logistic Accuracy:",ALP)
```

```python
# Plot confusion matrix
cm = confusion_matrix(y_test, y_LogisticPred)
sns.heatmap(cm, annot=True, fmt='g')
```

```python
# Classification Report
print(classification_report(y_test,y_LogisticPred))
```
